```python
from faulttools import draw, generate
from faulttools.web.partitions import pauli_webs_through_partitions

d, nodes = generate.shor_extraction(
    generate.rotated_planar_surface_code_stabilisers(3),
    qubits=3**2,
    repeat=2,
    granular=True,
    partition=True,
)

stabs, regions = pauli_webs_through_partitions(d, partitions=nodes)
for reg in regions:
    draw(d, web=reg)
```
